```python
from xmitgcm import llcreader
model = llcreader.ECCOPortalLLC4320Model()
ds = model.get_dataset(varnames=['Eta'], type='latlon')
```

```python
myds = ds.sel(time=slice('2012-01-01','2012-01-02'))
myds
```

```python
%%time
myds.to_netcdf('../DATA/LLC4320_eta.nc')
```
