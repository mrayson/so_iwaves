```python
N2_SO_Mar, zmid = gsw.stability.Nsquared(myS['s_an'][2,:], myT['t_an'][2,:], myT['depth'])
N2_SO_Aug, zmid = gsw.stability.Nsquared(myS['s_an'][7,:], myT['t_an'][7,:], myT['depth'])
N2_ITF_Mar, zmid = gsw.stability.Nsquared(myS2['s_an'][2,:], myT2['t_an'][2,:], myT['depth'])
N2_ITF_Aug, zmid = gsw.stability.Nsquared(myS2['s_an'][7,:], myT2['t_an'][7,:], myT['depth'])

plt.figure(figsize=(5,8))
plt.plot(N2_SO_Mar, -zmid,'b')
plt.plot(N2_SO_Aug, -zmid,'b:')
plt.plot(N2_ITF_Mar, -zmid,'orange')
plt.plot(N2_ITF_Aug, -zmid,'orange',ls=':')
# plt.plot(myS['s_an'][8,:], -myds['depth'],'b:')

# plt.plot(myS2['s_an'][3,:], -myds['depth'],'orange')
# plt.plot(myS2['s_an'][8,:], -myds['depth'],'orange',ls=':')

plt.legend(('SOTS - March','SOTS - August','ITF - March','ITF - August'))
plt.xlim(0, 5e-4)
plt.ylim(-1500,0)
plt.ylabel('Depth [m]')
plt.xlabel('$N^2$ [s$^{-2}$]')
plt.grid(b=True,ls=':')
plt.tight_layout()
plt.savefig('../FIGURES/VerticalStratification_N2.png', dpi=150)
```
